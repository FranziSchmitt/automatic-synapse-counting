```python
def grey_sequence(rgb_sequence):
    
    grey_images = []
    
    for image in rgb_sequence:
        grey_image = rgb2grey(image)
        grey_images.append(grey_image)
    
    return grey_images

grey_images = grey_sequence(images)
```
